```python
q3 = '''
    SELECT
        launch_angle,
        launch_speed,
        hit_distance_sc,
        COALESCE(atan((hc_x-125.42)/NULLIF((198.27-hc_y), 0))*180/PI()*.75, 67.5) AS spray_angle,
        events,
        hc_x,
        hc_y,
        CASE
            WHEN events IN ('single', 'double', 'triple', 'home_run') THEN 1
            ELSE 0
        END AS hit
    FROM statcast
    WHERE type='X' and game_year=2019 and events NOT IN ('field_error', 'fielders_choice');
'''

batted_balls = pd.read_sql_query(q3, conn, params=[pitcher])
batted_balls.head(15)
```

```python
q4 = '''
    SELECT
        home_team,
        away_team,
        inning_topbot,
        CASE
            WHEN inning_topbot = 'Bot' THEN away_team
            ELSE home_team
        END AS def_team
    FROM statcast
    WHERE type='X' and game_year=2019 and events NOT IN ('field_error', 'fielders_choice');
'''

batted_balls2 = pd.read_sql_query(q4, conn, params=[pitcher])
batted_balls2.head(15)
```

```python
q4 = '''
    SELECT *
    FROM statcast
    WHERE type='X' and game_year=2019 and events NOT IN ('field_error', 'fielders_choice')
    ORDER BY game_date, game_pk, at_bat_number;
'''

batted_balls2 = pd.read_sql_query(q4, conn, params=[pitcher])
batted_balls2.head(15)
```

```python
q5 = '''
WITH bballs19 AS
(
    SELECT
        game_date,
        game_pk,
        at_bat_number,
        home_team,
        launch_angle,
        launch_speed,
        hit_distance_sc,
        COALESCE(atan((hc_x-125.42)/NULLIF((198.27-hc_y), 0))*180/PI()*.75, 67.5) AS spray_angle,
        CASE
            WHEN events IN ('single', 'double', 'triple', 'home_run') THEN 1
            ELSE 0
        END AS hit,
        CASE
            WHEN inning_topbot = 'Top' THEN home_team
            ELSE away_team
        END AS def_team
    FROM statcast
    WHERE type='X' and game_year=2019 and events NOT IN ('field_error', 'fielders_choice')
)
SELECT
    home_team,
    launch_angle,
    launch_speed,
    hit_distance_sc,
    spray_angle,
    hit,
    avg(hit) OVER (PARTITION BY def_team ORDER BY game_date, game_pk, at_bat_number ROWS BETWEEN 149 PRECEDING AND CURRENT ROW) AS hit_rate_wk,
    avg(hit) OVER (PARTITION BY def_team ORDER BY game_date, game_pk, at_bat_number ROWS BETWEEN 749 PRECEDING AND CURRENT ROW) AS hit_rate_mo
    FROM bballs19;
'''

batted_balls3 = pd.read_sql_query(q5, conn)
batted_balls3.head(15)
```

```python
q6 = '''
    SELECT
        statcast.launch_angle,
        statcast.launch_speed,
        teams.mlb_id, 
        stadiums.*
    FROM stadiums
    JOIN teams ON stadiums.team = teams.team
    JOIN statcast ON teams.mlb_id = statcast.home_team
    WHERE statcast.type='X' AND game_year=2019 AND events NOT IN ('field_error', 'fielders_choice')
    ORDER BY statcast.game_date, statcast.game_pk, statcast.at_bat_number;
'''

stadiums = pd.read_sql_query(q6, conn)
stadiums.head(10)
```
